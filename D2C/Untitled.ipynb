```python
#Taking majority vote of all models on test data
majority_vote = [Counter([pred[i], pred[i],pred[i]]) for i in range(len(pred))]
y_pred = []
for counter in majority_vote:
 for k, v in counter.items():
   if v == max(counter.values()):
    y_pred.append(k)
```
